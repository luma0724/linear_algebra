**Linear combination**: 
For any scalars $x$ and $y$, vectors $v$ and $w$, there is a linear combination $xv + yw$ such that
$$xv + y\bb{w} = x\begin{bmatrix}a\\b\end{bmatrix} + y\begin{bmatrix}c\\d\end{bmatrix} = \begin{bmatrix}xa + yc\\xb + yd\end{bmatrix}$$

````python
import numpy as np
from random import randint

x = randint(0, 10)
y = randint(0, 10)
v = np.array([randint(0, 10) for _ in range(2)])
w = np.array([randint(0, 10) for _ in range(2)])
print(f'{x}{v} + {y}{w} = {x*v + y*w}')
````
